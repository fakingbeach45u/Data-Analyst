Peringatan DtypeWarning terjadi karena beberapa kolom pada dataframe memiliki jenis data yang bercampur (mixed types), yaitu terdapat lebih dari satu jenis data dalam satu kolom. Sebagai contoh, pada kolom 7,8, dan seterusnya memiliki lebih dari satu jenis data, misalnya float dan string. Hal ini dapat mengakibatkan kesalahan dalam analisis data karena jenis data yang tidak sesuai.

Untuk menangani hal ini, bisa dilakukan beberapa hal, di antaranya:

1. Menentukan jenis data pada setiap kolom saat membaca file CSV, misalnya:
df = pd.read_csv('nama_file.csv', dtype={'kolom1': 'jenis_data1', 'kolom2': 'jenis_data2', ...})

2. Mengubah jenis data pada kolom yang bercampur menjadi satu jenis data yang sesuai, misalnya:
df['kolom7'] = df['kolom7'].astype('float64')
df['kolom8'] = df['kolom8'].astype('float64')
dst

3. Mengabaikan peringatan DtypeWarning dengan menambahkan low_memory=False saat membaca file CSV, misalnya:
df = pd.read_csv('nama_file.csv', low_memory=False)

Namun, pilihan ketiga ini tidak dianjurkan jika ukuran file CSV cukup besar karena dapat mempengaruhi kinerja komputer dan memakan waktu yang lebih lama

# Merge The Table

```python
#Merge the table with inner join
df_merged = pd.merge(idn_df, popularity_df, how='inner', left_on='Uri', right_on='uri')
print(df_merged.shape)
df_merged.head(3)
```

```python
# Mengganti nilai pada kolom "Uri"
df_merged['Id'] = df_merged['Uri'].apply(lambda x: x.split('/')[-1])

# Menghapus kolom yang tidak diperlukan
df_merged.drop(['Uri', 'uri'], axis=1, inplace=True)
```

```python
# Memindahkan kolom "Id" ke kolom paling awal
list_cols = list(df_merged.columns)
list_cols = [list_cols[-1]] + list_cols[:-1]
df_merged = df_merged[list_cols]
print(df_merged.shape)
df_merged.head(3)
```

```python
# Ganti nama beberapa kolom
df_merged = df_merged.rename(columns={'country':'country_chart', 'date':'date_chart',
                                      'position':'position_chart'})
```

```python
# Check and drop duplikat jika ada
merged_dup = df_merged.drop_duplicates()
print('dimension of the data =', merged_dup.shape,'\n')
```

```python
df_merged.info()
```

```python
# Generate a correlation matrix
corr_matrix = df_merged.corr()

# Create a custom mask to hide the upper triangle and right side of the matrix
mask = np.zeros_like(corr_matrix)
mask[np.triu_indices_from(mask, k=1)] = True

# Set the figure size
plt.figure(figsize=(15, 7))

# Plot the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='RdYlGn', vmin=-1, vmax=1, mask=mask, center=0)

# Set the title and y-axis label rotation
plt.title('Correlation Heatmap', fontsize=14)
plt.yticks(rotation=0)

# Show the plot
plt.show()
```
